This dataset is highly unbalanced

0 -> Normal Transactions
1 -> Fraud Transactions

```python
#separating the data for analysis

legit= credit_card_data[credit_card_data.Class == 0]
fraud= credit_card_data[credit_card_data.Class == 1]
```

<h3>Under-Sampling</h3> as  284315 for legit transanctions whereas only 492 for fraudulent transanctions


Build a sample dataset containing similar distribution of normal transanctions and fraudulent transanctions

Number of Fraudulent Transanctions ---> 492

```python
legit_sample= legit.sample(n=492)
```

Concatenating two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class'].astype(int)
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training and Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression(max_iter=20000)
```

```python
#training the logistic regression model with training data

model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data: ',training_data_accuracy*100)
```

```python
#accuracy on testing data

X_test_prediction= model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on training data: ',test_data_accuracy*100)
```
